# ONE HOT ENCODING

One-Hot Encoding is another popular technique for treating categorical variables. It simply creates additional features based on the number of unique values in the categorical feature. Every unique value in the category will be added as a feature. One-Hot Encoding is the process of creating dummy variables. In this notebook, I have performed this technique on a mercedes dataset which consiste more than 350 categorical variables. I have taken few columns as an example and performed the one hot technique. 

```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mercedes_benz')
```

```python
df.shape
```

```python
df['X2'].value_counts()
```

Now, we will find the total number of distinct categories or labels present in each column. Later, we can replace them by 0 or 1 by one hot encoding technique.

```python
for item in df:
    print(item, ':', len(df[item].unique()), ' categories' )
```

Below, we can see that the 6 columns 'X1', 'X2','X3','X4', 'X5','X6 ' have been converted into 117 columns when we have performed the one hot emcoding.

```python
pd.get_dummies(df, drop_first= True).shape
```

```python
df['X2'].value_counts().shape
```

Here, we have used the technique of encoding which indeed takes the top 10 categories for a specific column, if those top 10 categories matches with any of them in the variable it is going to replace that specific category into 1 and the remaining categories to 0. Suppose let's consider 'X2' column we will be pulling out the top10 categories and compare those with the actual categories which are present in the 'X2' column and if you find any matching categories in those column just replace them with 1 orelse replace them with 0.

```python
top10 = [x for x in df.X2.value_counts().sort_values(ascending= False).head(10).index]
```

```python
top10
```

```python
for i in top10:
    df[i] = np.where(df['X2']== i, 1, 0)
    
df[['X2'] + top10].head(20)   
```

```python
def one_hot_encod(data, variables, top_x):
    for item in top_x:
        data[variables + '_' + item] = np.where(data[variables] == item, 1, 0)
        
one_hot_encod(df, 'X2', top10)
df.head()
```
